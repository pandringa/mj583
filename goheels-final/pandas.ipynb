```python
import json
import pandas as pd
```

```python
with open('heels_mbb.json', 'r') as jsonfile:
    data = json.loads(jsonfile.read())
data = [d for d in data if ('Stats' in d) and ('career_stats' in d['Stats']) and ('gamehigh_stats' in d['Stats']) and ('current_stats' in d['Stats'])]
names = [d['Full Name'] for d in data]
len(data)
```

```python
career_frames = [pd.DataFrame.from_dict(d['Stats']['career_stats']).apply(pd.to_numeric, errors='ignore') for d in data]
career_frames[0]
```

```python
gamehigh_frames = [pd.DataFrame.from_dict(d['Stats']['gamehigh_stats']).apply(pd.to_numeric, errors='ignore').transpose() for d in data]
gamehigh_frames[0]
```

```python
current_frames = [pd.DataFrame.from_dict(d['Stats']['current_stats']).apply(pd.to_numeric, errors='ignore').transpose() for d in data]
current_frames[0]
```

```python
# Compare Stats for the current season between players
player_totals = {}
for i, player in enumerate(data):
    player_totals[player['Full Name']] = current_frames[i].loc[' Total']
player_compare = pd.DataFrame.from_dict(player_totals,orient="index")
player_compare.drop('Opponent', axis=1)
```

```python
# See who holds the record for each stat
gamehighs = []
for i, df in enumerate(gamehigh_frames):
    df['Player'] = names[i]
    gamehighs.append( df.apply(pd.to_numeric, errors='ignore') )
frame = pd.concat(gamehighs)
idx = frame.groupby('Statistic')['Value'].transform(max) == frame['Value']
frame = frame[idx]
frame['Player'] = frame.groupby('Statistic')['Player'].transform(lambda name: ', '.join(name))
frame[['Value', 'Player']]
```
